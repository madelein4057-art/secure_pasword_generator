```python
import random
import string
```

Funcion para generar contraseñas

```python
def generar_contrasena(longitud, mayusculas, numeros, simbolos):
  caracteres = string.ascii_lowercase

  if mayusculas:
    caracteres += string.ascii_uppercase

  if numeros:
    caracteres += string.digits

  if simbolos:
    caracteres += string.punctuation

  contrasena = ""

  for i in range(longitud):
    contrasena += random.choice(caracteres)

  return contrasena
```

Evaluar fortaleza de la contraseña

```python
def evaluar_fortaleza(contrasena):

  longitud =len(contrasena)

  tiene_mayusculas = any(c.isupper() for c in contrasena)
  tiene_numeros = any(c.isdigit()for c in contrasena)
  tiene_simbolos = any(c in string.punctuation for c in contrasena)

  puntuacion = 0

  if longitud >= 8:
    puntuacion += 1

  if longitud >= 10:
    puntuacion += 1

  if tiene_mayusculas:
    puntuacion += 1

  if tiene_numeros:
    puntuacion += 1

  if tiene_simbolos:
    puntuacion +=1

  if puntuacion <= 2:
    return "Debil"

  elif puntuacion <= 4:
    return "Media"

  else:
    return "Fuerte"




```

Programa principal

Paso 1: Pedir longitud

```python
longitud = int(input("Ingrese la longitud de la contraseña:"))
```

Paso 2: Validar si es menor de 8

```python
if longitud < 8:
  print("Advertencia: la contrasena será débil.")

  continuar = input("Desea continuar? (si/no):")

  if continuar.lower() != "si":
    print("Programa finalizado")
```

Paso 3: pedir las preferencias

```python
mayusculas = input("Incluir mayúsculas? (si/no):").lower() == "si"

numeros = input("Incluir numeros?(si/no):").lower() == "si"

simbolos = input("Incluir simbolos? (si/no):").lower() == "si"
```

Paso 4: Generar contraseña

```python
contrasena = generar_contrasena(
    longitud,
    mayusculas,
    numeros,
    simbolos
)
```

paso 5: Mostrar contraseñas

```python
print("\nContraseña generada:")
print(contrasena)
```

Paso 6: evaluar fortaleza

```python
fortaleza = evaluar_fortaleza(contrasena)
print("\nFortaleza:")
print(fortaleza)
```

Paso 7: Guarda contraseña

```python
guardar = input("\n¿Desea guardar la contraseña? (si/no): ")
```

Paso 8: Crear archivo TXT

```python
if guardar.lower() == "si":
  archivo = open("contrasena_guardada", "w")

  archivo.write(contrasena)

  archivo.close()

  print("Contraseña guardada correctamente en contrasenas_guardadas")

else:
  print("La contraseña no fue guardada")
```
